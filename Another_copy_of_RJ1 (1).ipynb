```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(32, activation='relu',input_shape=(273,)))
model.add(tf.keras.layers.Dense(64, activation='relu'))
model.add(tf.keras.layers.Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

```python
from sklearn.metrics import matthews_corrcoef,f1_score
import tensorflow as tf

# Definisanje metrika
def mcc_metric(y_true, y_pred):
    y_pred_rounded = tf.round(y_pred)
    return matthews_corrcoef(y_true, y_pred_rounded.numpy())

def f1_metric(y_true, y_pred):
    y_pred_rounded = tf.round(y_pred)
    return f1_score(y_true, y_pred_rounded.numpy())

```

```python
model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy', mcc_metric, f1_metric, tf.keras.metrics.AUC()],
              run_eagerly=True)
```

```python
X_train.shape
```

```python
povijest = model.fit(X_train,y_train, epochs=10, batch_size=128)
```

```python
model.evaluate(X_test, y_test)
```

regularizacija

```python
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras import regularizers

model3 = tf.keras.Sequential()
model3.add(tf.keras.layers.Dense(64, activation='relu',input_shape=(273,),kernel_regularizer=regularizers.l2(0.01)))
model3.add(Dropout(0.5))
model3.add(tf.keras.layers.Dense(32, activation='relu',kernel_regularizer=regularizers.l2(0.01)))
model3.add(Dropout(0.5))

model3.add(tf.keras.layers.Dense(64, activation='relu',kernel_regularizer=regularizers.l2(0.01)))
model3.add(Dropout(0.5))

model3.add(tf.keras.layers.Dense(1, activation='sigmoid'))
```

```python
model3.summary()
```

```python
model3.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy', mcc_metric, f1_metric, tf.keras.metrics.AUC()],
              run_eagerly=True)
```

```python
povijest = model3.fit(X_train,y_train, epochs=10, batch_size=128)
```

```python
model3.evaluate(X_test, y_test)
```
